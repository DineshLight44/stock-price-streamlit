```python
import pandas as pd 
import numpy as np 
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
```

```python
data = pd.read_csv("merge-csv.com__684d1e1110c5f.csv", skiprows=2, on_bad_lines='skip',low_memory=False)
data.shape
```

```python
stock_data = data[data['Symbol'] == 'RELIANCE']
```

```python
print(data.columns)
```

```python
stock_data = stock_data.sort_values('Date')
```

```python
close_prices = stock_data['Close'].values.reshape(-1, 1)
```

```python
scaler = MinMaxScaler(feature_range=(0,1))
scaled_prices = scaler.fit_transform(close_prices)
```

```python
def create_sequences(data, seq_length):
    X, y = [], []
    for i in range(len(data) - seq_length):
        X.append(data[i:i+seq_length])
        y.append(data[i+seq_length])
    return np.array(X), np.array(y)
```

```python
seq_length = 60
X, y = create_sequences(scaled_prices, seq_length)
```

```python
split = int(0.8*len(X))
X_train, X_test = X[:split], X[split:]
y_train , y_test = y[:split] , y[split:]
```

```python
print("Training shape",X_train.shape, y_train.shape)
print("Testing shape",X_test.shape, y_test.shape)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
```

```python
model = Sequential()
model.add(LSTM(100, return_sequences=True, input_shape=(seq_length, 1)))

model.add(LSTM(50))
model.add(Dense(1))

model.compile(optimizer = 'adam',loss= 'mean_squared_error')

history = model.fit(X_train, y_train, epochs=50, batch_size=32, validation_data=(X_test, y_test))
```

```python
pred_prices = model.predict(X_test)

pred_prices = scaler.inverse_transform(pred_prices)
real_price = scaler.inverse_transform(y_test.reshape(-1,1))
```

```python
plt.figure(figsize=(12,6))
plt.plot(real_price, color='blue', label='Actual Prices')
plt.plot(pred_prices, color='red', label='Predicted Prices')
plt.title('Stock Close Price Prediction')
plt.xlabel('Time')
plt.ylabel('Price')
plt.legend()
plt.show()
```

```python
model.save('stock_rnn_model.h5')
```
